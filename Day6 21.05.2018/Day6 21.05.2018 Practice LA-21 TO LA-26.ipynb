```python
#LA 26
#Write a code to calculate income tax of people.
#taxfile1.txt contains name and income of users of 1st month.
#Similar files exist for next 2 months with name order changed.
#rate.txt contains tax rates as follows:
#upto Rs2,50,500 no Tax
#Rs. 2,50,500 - Rs. 5,00,000 5%
#Rs. 5,00,000 - Rs. 10,00,000 20%
#Rs. 10,00,000 and beyond 30%
f1=open("taxfile1.txt","r")
f2=open("taxfile2.txt","r")
f3=open("taxfile3.txt","r")
dict1={}
for i in [f1,f2,f3]:
    cust=i.readlines()
    for j in cust:
        j=j.split()
        if(j[0] in dict1):
            dict1[j[0]]=dict1[j[0]]+float(j[1])
        else:
            dict1[j[0]]=float(j[1])
    #print(dict1)
print(dict1)
dict2={}
rt=open("rate.txt","r")
rate=rt.readlines()
for i in dict1:
    for j in rate:
        j=j.split()
        if(dict1[i]>=float(j[0]) and dict1[i]<=float(j[1])):
            tax_rate=float(j[2])
            dict2[i]=tax_rate*dict1[i]*0.01
print(dict2)
for i in dict2:
    print("User {0} Tax: {1}".format(i,dict2[i]))
f1.close()
f2.close()
f3.close()0
```
