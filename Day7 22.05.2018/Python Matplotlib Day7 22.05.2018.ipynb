```python
#imports needed for function
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x=np.linspace(0,5,10) #10 samplew points from 0 to 5
print(x) #will print the array
print(type(x)) #numpy array
y=x**2 #the equation y is raised to 2 power of x
```

```python
#why this? output comes without this
figure()
```

```python
plot(x,y,'r') #plot x wrt y and 'r'-red
xlabel('x')
ylabel('y')
title('title')
show() #mandatory without this it wont show output
```

```python
#subplot
subplot(1,2,1) #1 row 2 columns 1st one
plot(x,y,'r--') #y wrt x and r-- means red line in dashes
subplot(1,2,2) #1 row 2 column 2nd one
plot(y,x,'g*-') #x wrt y *- green line
```

```python
#create x sample containing 150 points from 5 to 50
#then plot the equations
#y=3*x+7 plot with amgenta plus style m+
#x=(y-3)/3 plot with blue style b--
#plot both side by side 

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

x=np.linspace(5,50,150)
y=(x*3)+7
x1=(y-3)/3
figure()
xlabel('x')
ylabel('y')
title('title')
subplot(2,2,1) #1 row 2 columns 1st one
plot(x,y,'m+') #y wrt x and r-- means red line in dashes
subplot(2,2,2) #1 row 2 column 2nd one
plot(y,x1,'g--') #x wrt y *- green line
show()
```
